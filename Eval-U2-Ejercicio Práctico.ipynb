```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

# Generar datos sintéticos
X, y = make_classification(n_samples=100, n_features=20, n_informative=10, n_classes=2, random_state=42)

# Modelo base
scores_base_rf = []
for i in range(10):
    clf = RandomForestClassifier(random_state=42)
    clf.fit(X, y)
    scores_base_rf.append(clf.score(X, y))

# Modelo con hiperparámetros
clf = RandomForestClassifier(n_estimators=100, criterion="gini", max_depth=None, min_samples_split=2, min_samples_leaf=1, max_features=12, random_state=42)
scores_hiper_rf = []
for i in range(10):
    clf.fit(X, y)
    scores_hiper_rf.append(clf.score(X, y))

# Boxplot
fig, ax = plt.subplots()
bp_dict = ax.boxplot([scores_base_rf, scores_hiper_rf], labels=['Base', 'Hiperparámetros'])
ax.set_title('Random Forest')
plt.show()
```

```python
from sklearn.svm import LinearSVC
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt

# Generar datos sintéticos
X, y = make_classification(n_samples=100, n_features=20, n_informative=10, n_classes=2, random_state=42)

# Modelo base
scores_base_svc = []
for i in range(10):
    clf = LinearSVC(random_state=42)
    clf.fit(X, y)
    scores_base_svc.append(clf.score(X, y))

# Modelo con hiperparámetros
clf = LinearSVC(C=0.1, penalty='l1', dual=False, random_state=42)
scores_hiper_svc = []
for i in range(10):
    clf.fit(X, y)
    scores_hiper_svc.append(clf.score(X, y))

# Boxplot
fig, ax = plt.subplots()
bp_dict = ax.boxplot([scores_base_svc, scores_hiper_svc], labels=['Base', 'Hiperparámetros'])
ax.set_title('Linear Support Vector Classifier')
plt.show()
```

```python
print("Promedio de scores base (Random Forest):", np.mean(scores_base_rf))
print("Promedio de scores hiperparámetros (Random Forest):", np.mean(scores_hiper_rf))
print("Promedio de scores base (Linear Support Vector Classifier):", np.mean(scores_base_svc))
print("Promedio de scores hiperparámetros (Linear Support Vector Classifier):", np.mean(scores_hiper_svc))
```

```python
 #comparar los promedios de los scores
if np.mean(scores_base_rf) > np.mean(scores_base_svc) and np.mean(scores_hiper_rf) > np.mean(scores_hiper_svc):
    print("El modelo de Random Forest obtuvo mejores scores en ambos casos, tanto en la configuración base como en la de hiperparámetros.")
else:
    print("El modelo de Linear Support Vector Classifier obtuvo mejores scores en al menos uno de los casos, o ambos modelos tuvieron scores similares.")
```
